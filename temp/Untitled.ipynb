```python
import pandas as pd
import numpy as np
import json, os
import pyarrow as pq
```

```python
def json_save(file_path, data):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)
```

```python
def json_load(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)
```

```python
def create_metrics_overview():
    container = set([])
    n_rows = 0
    
    for dir in os.listdir('1702372324'):
        metrics = json_load(f'1702372324/{dir}/metrics.json')
        n_rows += len(metrics)

        for metric in metrics:
            keys = list(metric.keys())
            container = container.union(keys)

    columns = list(container)
    columns.sort()
    n_cols = len(columns)

    print(n_cols, n_rows)

    ###################################################

    data = np.full((n_rows, n_cols), 'TEMP')
    df = pd.DataFrame(data, columns=columns)
    next_row = 0

    for dir in os.listdir('1702372324'):
        metrics = json_load(f'1702372324/{dir}/metrics.json')

        for metric in metrics:
            df.loc[next_row] = metric
            next_row += 1

    df.rename(columns={'__name__':'metric_name'}, inplace=True)
    
    return df
```

```python
df = create_metrics_overview()
```

```python
df.to_csv('metrics_overview.csv')
```
